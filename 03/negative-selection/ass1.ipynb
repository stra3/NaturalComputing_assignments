```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
unsorted = open('unsorted.test')
lines = unsorted.read().split(' \n')
numbers = [float(a) for a  in lines[:-1]] #for some reason the last element of lines was a space


#unsorted = english + tagalog 
#english length = 124
#tagalog length = 708
#total   length = 832

#labeling data: 0=english, 1=tagalog

scores = np.zeros((832,2))
for i in range(len(scores)):
    scores[i][0] = numbers[i]
    if i<132:
        scores[i][1] = 0
    else: 
        scores[i][1] = 1 

scores = list(reversed(sorted(scores, key=lambda x: x[0])))
#scores = sorted(scores, key=lambda x: x[0])
```

```python
sensitivity = []

for i in range(len(scores)):
    total = 0
    for j in range(i+1, len(scores)):
        if scores[j][1] == 1:
            total+=1
    sensitivity.append(total/(708))
    
specificity = []

for i in range(len(scores)):
    total = 0
    for j in range(i-1, 0, -1):
        if scores[j][1] == 0:
            total+=1
    specificity.append(total/(124))
```

```python
print(max(sensitivity))
```

```python
plt.plot(specificity,sensitivity[::-1])
```

```python
text_file = open("sorted.test")

lines = text_file.read().split(' \n')
numbers = [float(a) for a  in lines[:-1]]
```

```python
print(numbers)
```

```python
sensitivity = []
total = len(numbers)
for n in range(1, len(numbers)+1):
    sensitivity.append(n/total)
```

```python
specificity = []
for n in range(1,len(numbers)+1):
    specificity.append( (n-1)/total)
```
