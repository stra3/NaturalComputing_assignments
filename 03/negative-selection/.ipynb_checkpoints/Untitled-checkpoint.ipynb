```python
import numpy as np
```

```python
text_file = open("sorted.test")

lines = text_file.read().split(' \n')
numbers = [float(a) for a  in lines[:-1]]
```

```python
print(numbers)
```

```python
sensitivity = []
total = len(numbers)
for n in range(1, len(numbers)+1):
    sensitivity.append(n/total)
```

```python
print(sensitivity)
```
